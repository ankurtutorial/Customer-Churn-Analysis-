```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("C:\\Users\\MANISH SINGH\\Favorites\\Downloads\\customer_churn (1).csv")

# Calculate counts for each category
counts = df.groupby(['SeniorCitizen', 'Churn']).size().unstack(fill_value=0)

# Calculate the percentage for each category within SeniorCitizen group
percentage = counts.div(counts.sum(axis=1), axis=0) * 100

# Plot the stacked bar chart
fig, ax = plt.subplots(figsize=(5, 5))
percentage.plot(kind='bar', stacked=True, color=["lightblue", "salmon"], ax=ax)

# Add labels and title
plt.title("Churn by SeniorCitizen (Stacked bar chart)")
plt.xlabel("Senior Citizen (1: Yes, 0: No)")
plt.ylabel("Percentage")
plt.xticks(rotation=0)

# Correctly position percentage labels on bars
for i, (index, row) in enumerate(percentage.iterrows()):
    no_churn = row[0]  # No churn percentage
    yes_churn = row[1]  # Yes churn percentage
    
    ax.text(i, no_churn / 2, f"{no_churn:.1f}%", ha='center', va='center', fontsize=10, color='black')
    ax.text(i, no_churn + yes_churn / 2, f"{yes_churn:.1f}%", ha='center', va='center', fontsize=10, color='black')

# Show legend
plt.legend(title="Churn", labels=["No", "Yes"])
plt.show()
```

#Comparative a greated pecentage of people in senior citizen category have churned

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data = {
    'PhoneService': ['Yes', 'No', 'Yes', 'Yes', 'No', 'Yes', 'No'],
    'MultipleLines': ['No', 'Yes', 'No', 'Yes', 'No', 'No', 'Yes'],
    'InternetService': ['DSL', 'Fiber optic', 'DSL', 'Fiber optic', 'No', 'DSL', 'No'],
    'OnlineSecurity': ['No', 'Yes', 'No', 'No', 'Yes', 'No', 'Yes'],
    'OnlineBackup': ['Yes', 'No', 'Yes', 'No', 'Yes', 'Yes', 'No'],
    'DeviceProtection': ['No', 'No', 'Yes', 'No', 'Yes', 'No', 'Yes'],
    'TechSupport': ['Yes', 'No', 'Yes', 'No', 'Yes', 'No', 'No'],
    'StreamingTV': ['No', 'Yes', 'No', 'No', 'Yes', 'Yes', 'No'],
    'StreamingMovies': ['Yes', 'No', 'Yes', 'No', 'Yes', 'No', 'No'],
    'Churn': ['Yes', 'No', 'Yes', 'No', 'No', 'Yes', 'No'],  }

df = pd.DataFrame(data)

fig, axes = plt.subplots(3, 3, figsize=(15, 12)) 
fig.suptitle("Count Plots for Different Services (Hue = Churn)", fontsize=16)

columns = df.columns[:-1]  
for ax, col in zip(axes.flatten(), columns):
    sns.countplot(x=df[col], hue=df['Churn'], ax=ax, palette="pastel")
    ax.set_title(col)
    ax.set_xlabel("")
    ax.set_ylabel("Count")
    ax.tick_params(axis='x', rotation=45)

plt.tight_layout(rect=[0, 0, 1, 0.96]) 
plt.show()
```

```python
#InternetService, and OnlineSecurity have higher retention, with more customers staying. StreamingTV, StreamingMovies,and 
#DeviceProtection show more balanced churn, with a significant number of customers leaving.Churn rates vary widely depending
#on the service.
```
